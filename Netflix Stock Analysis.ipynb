Netflix Stock Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data=pd.read_csv(r"C:\Users\hp\Downloads\Netflix.csv")
data.head()
```

```python
data.set_index('Date',inplace=True)
```

Creating new Column and extracting year from Date

```python
data['Date']=pd.to_datetime(data['Date'])
data['year']=data['Date'].dt.year
data['year'].astype(int)
```

Statement 1: Volume Of Stock Traded 

```python
sns.lineplot(x='year',y='Volume',data=data,label='Volume')
plt.title('Volume Of Stock Vs Year')
plt.xticks(data['year'])
plt.show()
```

Statement 2:Netflix Stock Price-High,Open,Close

```python
data.plot(y=['High','Close','Open'],title="Netflix Stock price")
```

Netflix Stock Price-Day,Month,Year -Wise

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(15, 10))
data.groupby(data.index.day).mean().plot(y='Volume',ax=ax1,xlabel='Day')
data.groupby(data.index.month).mean().plot(y='Volume',ax=ax2,xlabel='Month')
data.groupby(data.index.year).mean().plot(y='Volume',ax=ax3,xlabel='Year')
```

Statement 4:Top 5 Dates With Highest Stock price

```python
highest=data.sort_values(by='High',ascending=False).head(5)
highest['High']
```

Statement 5 :Top 5 Dates With Lowest Stock price

```python
lowest=data.sort_values(by='Low',ascending=True).head(5)
lowest['Low']
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(12,5),sharex=True)
fig.suptitle("High And Low value Stock Per Date",fontsize=18)
sns.lineplot(ax=axes[0],y=data['High'],x=data.index,color='green')
sns.lineplot(ax=axes[1],y=data['Low'],x=data.index,color='red')
```
